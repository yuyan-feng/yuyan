```python
img_message = """
"""
video_message = """
</ul>
</div>

"""
video_num = 1
for file in files:
    if 'jpg' in file:
        jpg_time = str(file)[0:10].replace('_', '年', 1).replace('_', '月') + '日'
        print(jpg_time)
        img_message += '    <li class="grid-item" data-src="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/%s/%s" data-sub-html="%s">\n'%(folder, file, jpg_time)
        img_message += '        <img src="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/%s/thumbnail/%s" class="img-thumbnail"/> </li>\n'%(folder, file)
    if 'mp4' in file:
        file_jpg = file.replace('mp4', 'jpg')
        img_message += '    <li class="grid-item video" data-poster="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/%s/thumbnail/%s" data-sub-html="" data-html="#video%s">\n'%(folder, file_jpg, video_num)
        img_message += '        <a href=""><img class="img-responsive" src="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/%s/%s">\n'%(folder, file_jpg)
        img_message += '            <div class="demo-gallery-poster"><img src="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/play-button.png"></div></a></li>\n'
        video_message += '    <div style="display:none;" id="video%s">'%video_num
        video_message += '        <video class="lg-video-object lg-html5 video-js vjs-default-skin" controls>\n'
        video_message += '            <source src="https://cdn.jsdelivr.net/gh/yuyan-feng/yuyan/img/%s/%s" type="video/mp4">Your browser does not support HTML5 video.</video></div>\n'%(folder, file)
        video_num += 1
```
